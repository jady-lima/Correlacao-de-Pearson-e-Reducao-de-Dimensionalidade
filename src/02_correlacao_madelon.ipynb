```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
databaseMadelon = pd.read_csv("data/Madelon.csv", encoding="utf-8")
databaseMadelon.info()
```

```python
correlationPearson = databaseMadelon.iloc[:, 1:-1].corr(method='pearson')
print(correlationPearson)
```

```python
databaseMadelon = databaseMadelon.dropna()
```

```python
correlationPearson = databaseMadelon.corr(method='pearson')
print(correlationPearson)
```

```python
plt.figure(figsize=(25, 10))
plot = sn.heatmap(correlationPearson, annot=True, fmt=".1f", linewidths=.6)
plt.show()
```
