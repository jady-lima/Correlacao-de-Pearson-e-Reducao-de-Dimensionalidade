```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
databaseMadelon = pd.read_csv("data/Madelon/Madelon.csv", encoding="utf-8")
databaseMadelon.info()
```

```python
X = databaseMadelon.iloc[:, :-1]
correlationPearsonComplete = X.corr(method='pearson')
print(correlationPearsonComplete)
```

```python
df1 = pd.DataFrame(correlationPearsonComplete)
df1.to_csv("data/Madelon/CorrelacaoPearsonComplete.csv", index=False)
```

```python
plt.figure(figsize=(15, 8))
plot = sn.heatmap(correlationPearsonComplete, annot=True, fmt=".2f", linewidths=.6)
plt.show()
```

```python
correlationPearson100 = databaseMadelon.iloc[:100, :100].corr(method='pearson')
print(correlationPearson100)
```

```python
df1 = pd.DataFrame(correlationPearson100)
df1.to_csv("data/Madelon/CorrelacaoPearson100.csv", index=False)
```

```python
plt.figure(figsize=(25, 10))
plot2 = sn.heatmap(correlationPearson100, annot=True, fmt=".1f", linewidths=.6)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_std = StandardScaler().fit_transform(X)
print(X_std)
print(X_std.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca90 = PCA(n_components=0.90, whiten=True)
X_std_pca90 = pca90.fit_transform(X_std)

print(X_std_pca90)
print(X_std_pca90.shape)
```

```python
df1 = pd.DataFrame(X_std_pca90)
df1.to_csv("data/Madelon/PCA90.csv", index=False)
```

```python
pca10 = PCA(n_components=10, whiten=True)
X_std_pca10 = pca10.fit_transform(X_std)

print(X_std_pca10)
print(X_std_pca10.shape)
```

```python
df1 = pd.DataFrame(X_std_pca10)
df1.to_csv("data/Madelon/PCA10.csv", index=False)
```
